```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('taxis')
df
```

```python
sns.barplot(x='payment',y='total',data=df,color='red',estimator='std')
plt.show()
plt.tight_layout()
```

```python
sns.histplot(x=df['fare'],color='blue')
plt.show()
```

```python
sns.histplot(y=df['fare'],color='blue')
plt.show()
```

```python
sns.histplot(y=df['fare'],palette=['blue','red'],hue=df['payment'])
plt.show()
```

```python
sns.histplot(x=df['fare'],palette=['blue','red'],hue=df['payment'])
plt.show()
```

```python
sns.histplot(y=df['fare'],palette=['b','y'],hue=df['payment'])
plt.show()
```

```python
sns.histplot(x=df['fare'],palette=['r','y'],hue=df['payment'])
plt.show()
```

```python
sns.histplot(x=df['fare'],palette=['g','y'],hue=df['payment'])
plt.show()
```

```python
sns.distplot(x=df['fare'],hist=True,kde=False)
plt.show()
```

```python
sns.distplot(x=df['fare'],hist=True,kde=True)
plt.show()
```

```python
sns.distplot(x=df['fare'],hist=False,kde=False)
plt.show()
```

```python
sns.distplot(x=df['fare'],hist=True,kde=True)
plt.show()
```

```python
sns.boxplot(x=df['fare'],color='black')
plt.show()
```
